```python
def between(price):#0-10000
    return int(price) % 10000 + 0
```

```python
prices = list(map(between, prices))

prices
```

```python
try:
    connection = mysql.connector.connect(
        host=host,
        port=3306,
        user=user,
        passwd=password,
        database=db_name
    )
    
    print("successfuly connected...")
    try:
        with connection.cursor() as cursor:
            
#             createADatabase = "CREATE DATABASE BooksAndAuthors"
#             cursor.execute(createADatabase)

#             createTableQuery = "CREATE TABLE Authors (AuthID INT AUTO_INCREMENT PRIMARY KEY, Auth_Name VARCHAR(15), Surname VARCHAR(15))"
#             cursor.execute(createTableQuery)
            
            for auth in authors:
        
                name = auth["name"]
                surname = auth["surname"]
                
                #checking if this author is already into the table
                doesRowExist = "SELECT COUNT(*) " \
                                "FROM `Authors` " \
                                f"WHERE `Auth_Name`='{name}' AND `Surname`='{surname}'"
                
                cursor.execute(doesRowExist)
                amountExistingRow = cursor.fetchone()[0]

                if amountExistingRow == 0:
                    insertQuery = "INSERT INTO `Authors` " \
                                    "(`Auth_Name`, `Surname`)" \
                                    " VALUES " \
                                    f"('{name}', '{surname}')"
                    
                    cursor.execute(insertQuery)
                    
            connection.commit()
            print("The authors were append into the 'Authors' table")
            
            
            querySelect = "SELECT * FROM `Authors`"
            cursor.execute(querySelect)
            
            authorsById = []
            for auth in cursor:
                authorsById.append(auth[0])
                
            #Continue to generate. Namely, I drawed every authors out of the table 'Authors' to assign every generated book
            #its author.
            #Of cource, data are recieved from different APIs and a problem for data engineers doesn't consists of uniting
            #recieved data and data which are based on an already existing database. Opositly, A data engineer work consists of cleanng
            #recieved data and appending to a specific table from database
            #P.S.: it is things how I understand it
            
            chosenAuthorsById = []
            ranArray = random.sample(range(len(authorsById)), amount)
            
            for idx in ranArray:
                chosenAuthorsById.append(authorsById[idx])
                
            #end of the generating

#             createTableQuery = "CREATE TABLE `Books` (AuthID INT KEY, Title VARCHAR(50), Price FLOAT)"
#             cursor.execute(createTableQuery)

#             uniteByKey = "ALTER TABLE `booksandauthors`.`Books` " \
#                         "ADD CONSTRAINT `AuthID` " \
#                         "FOREIGN KEY (`AuthID`) " \
#                         "REFERENCES `booksandauthors`.`Authors` (`AuthID`) " \
#                         "ON DELETE NO ACTION " \
#                         "ON UPDATE NO ACTION"
#              cursor.execute(uniteByKey)

            for i in range(amount):
                idx = chosenAuthorsById[i]
                title = bookstitles[i]
                price = prices[i]
                insertQuery = "INSERT INTO `Books` " \
                                    "(`AuthID`, `Title`, `Price`)" \
                                    " VALUES " \
                                    f"({idx}, '{title}', {price})"
                
                cursor.execute(insertQuery)
            
            
            connection.commit()
            print("The authors were append into the 'Books' table")
                
                
            #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
            #practice
            
            print("\nPRACTICE\n")
            
            
            print("1.)Number of records in total\n")
            allAuthors = "SELECT COUNT(*) " \
                                "FROM `Authors`"
            cursor.execute(allAuthors)
            recordNumber = cursor.fetchone()[0]
            print(f"Number of authors is {recordNumber}\n")
            
            print("2.)Number of unique groups in the dataset\n")
            selectDistinctQuery = "SELECT `Title`, COUNT(`Title`) FROM `Books` GROUP BY `Title`"
            cursor.execute(selectDistinctQuery)
            uniqueBooks = cursor.fetchall()
            for title in uniqueBooks:
                print(f"The book quantity of '{title[0]}' equals to {title[1]} (it might be different author)")
            print("\n")
            
            
            print("3.)Number of unique groups in the dataset\n") 
            everyAuthor = "SELECT `AuthID`, `Auth_Name` FROM `Authors`"
            cursor.execute(everyAuthor)
            everyAuthor = cursor.fetchall()
            for auth in everyAuthor:
                idx = auth[0]
                quantityAuthorsBooks = "SELECT COUNT(*) FROM " \
                "(SELECT * FROM " \
                "(SELECT authors.AuthID, authors.Auth_Name, authors.Surname, books.Title, books.Price " \
                "FROM authors " \
                "INNER JOIN books ON authors.AuthID=books.AuthID) AS UNITED " \
                f"WHERE UNITED.AuthID = {idx} " \
                "GROUP BY UNITED.Title) AS CLEANED"
                cursor.execute(quantityAuthorsBooks)
                bookQuantity = cursor.fetchone()[0]
                AuthName = auth[1]
                print(f"{AuthName}'s books quantity equels to {bookQuantity}")
            print("\n")
            
            
            print("4.)Finding average price and time execution it\n")
            start_time = time.time()
            avaregeQuery = "SELECT AVG(`Price`) FROM `Books`"
            cursor.execute(avaregeQuery)
            avarege = cursor.fetchone()[0]
            print(f"The execution of finding the avarege price of all books in {time.time() - start_time}s and it equals to {avarege}")
    finally:
        connection.close();
except Exception as ex:
    print("Connection refused...")
    print(ex)
```
